# Regression with Heteroscedasticidy and Correlated Errors

```python
x_train = np.random.sample(10000)
y_train = np.add(5*x_train, np.multiply( (x_train)**2, np.random.standard_normal(x_train.shape)))

x_test = np.random.sample(10000)
y_test = np.add(5*x_test, np.multiply( (x_test)**2, np.random.standard_normal(x_test.shape)))


x_train = x_train[:, np.newaxis]
x_test = x_test[:, np.newaxis]
y_train = y_train[:, np.newaxis]
y_test = y_test[:, np.newaxis]

plt.figure(figsize=(20, 10))

plt.title(r'$y = 5x + x^2*\epsilon\backsim \mathit{N}(0,1)$', fontsize=32)

plt.plot(x_train,y_train, "x",alpha=1., color=tableau20[0])
plt.show()
```

```python
s = np.linspace(0.,1.,int(1e3))[:, np.newaxis]
#alpha_pred, mu_pred, sigma_pred = mdn.predict(s)
```

```python
from keras import backend as K
import tensorflow as tf
K.clear_session()

from keras.layers import Lambda, Reshape, Multiply, Activation
from keras.layers import Input, Dense, Dropout, Concatenate
from keras.layers.normalization import BatchNormalization

from keras.callbacks import EarlyStopping
from keras import regularizers

from keras.models import Model
from keras import optimizers

from keras.utils.generic_utils import get_custom_objects

epsilon = 1e-8

def elu1p(x):
    return (K.elu(x) + 1)+epsilon

def gaussian_likelihood(input, comps):
    alpha = comps[:,0:mixtures]
    mu = comps[:,mixtures:2*mixtures]
    sigma = comps[:,2*mixtures:3*mixtures]

    dist = tf.distributions.Normal(loc=mu, scale=sigma)

    prob = dist.prob(input)
    loglike = tf.multiply(prob, alpha)
    summ = tf.reduce_sum(loglike, axis=1, keep_dims=True)
    logloss = tf.reduce_mean(-tf.log(summ+epsilon))
    return logloss
    
def split_prediction(comps):
    alpha = comps[:,0:mixtures]
    mu = comps[:,mixtures:2*mixtures]
    sigma = comps[:,2*mixtures:3*mixtures]
    
    return alpha, mu, sigma
    
get_custom_objects().update({'elu1p': Activation(elu1p)})

neurons = 500
mixtures = 4

inputs = Input(shape=(x_train.shape[1],))

h1 = Dense(neurons, activation="relu", kernel_regularizer=regularizers.l2(1e-7))(inputs)
h2 = Dense(neurons, activation="relu", kernel_regularizer=regularizers.l2(1e-7))(h1)

mus = Dense(mixtures, name="mus")(h2)
sigmas = Dense(mixtures, activation="elu1p", name="sigmas")(h2)
alphas = Dense(mixtures, activation="softmax", name="alphas")(h2)
output = Concatenate(name="output")([alphas, mus, sigmas])

opt = optimizers.RMSprop(lr=0.0001, rho=0.9, decay=0.0, clipvalue=0.5)

mdn = Model(inputs=inputs, outputs=output)
mdn.compile(loss=gaussian_likelihood, optimizer=opt)


mon = EarlyStopping(monitor='val_loss', min_delta=0, patience=5, verbose=0, mode='min')
```

```python
mdn.fit(x=x_train, y=y_train,epochs=200, validation_data=(x_test, y_test), callbacks=[mon], batch_size=128, verbose=1)
y_pred = mdn.predict(s)
alpha_pred, mu_pred, sigma_pred = split_prediction(y_pred)
```

```python
plt.figure(figsize=(20, 10))

#plt.title('MDN', fontsize=48, fontname='Times New Roman')
plt.title(r'$y = 5x + x^2*\epsilon\backsim \mathit{N}(0,1)$', fontsize=32)

plt.plot(x_train,y_train,'x',alpha=.2, color="black", markersize=2)

for mx in range(mixtures):

    for i in range(len(mu_pred)):
        plt.errorbar(s[i],mu_pred[i,mx],
                        yerr=sigma_pred[i,mx],
                        alpha=alpha_pred[i,mx], 
                        color=tableau20[mx])
        
        plt.plot(s[i],mu_pred[i,mx], color=tableau20[mx],linewidth=1, marker='.', 
                linestyle='',alpha=alpha_pred[i,mx], markersize=10)# label='gaus_'+str(mx)+" prediction", )
        
ax = plt.gca()

ax.set_ylim(0, 7)
ax.set_xlim(0, 1)


plt.tight_layout()
plt.show()
```

# Non-Linear Regression with Multimodal Output

```python
def unison_shuffled_copies(a, b):
    assert len(a) == len(b)
    p = np.random.permutation(len(a))
    return a[p], b[p]


NSAMPLE = 10000
x_data = np.float32(np.random.uniform(-10.5, 10.5, (1, NSAMPLE)))
r_data = np.array([np.random.normal(scale=np.abs(i)) for i in x_data])
y_data = np.float32(np.square(x_data)+r_data*2.0)

x_data2 = np.float32(np.random.uniform(-10.5, 10.5, (1, NSAMPLE)))
r_data2 = np.array([np.random.normal(scale=np.abs(i)) for i in x_data2])
y_data2 = np.float32(-np.square(x_data2)+r_data2*2.0)

x_data = np.concatenate((x_data,x_data2),axis=1).T
y_data = np.concatenate((y_data,y_data2),axis=1).T

x_data,y_data = unison_shuffled_copies(x_data,y_data)

#min_max_scaler = MinMaxScaler()
#y_data = min_max_scaler.fit_transform(y_data)

plt.figure(figsize=(20, 15))
plot_out = plt.plot(x_data,y_data,"x",alpha=1., color=tableau20[0])

plt.title(r'$y = \pm x^2 + \epsilon$', fontsize=32)

plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.5, random_state=42)

s = np.linspace(-10.,10.,int(5e2))[:, np.newaxis]
mdn.reset_states()
mdn.fit(x=x_train, y=y_train,epochs=200, validation_data=(x_test, y_test), callbacks=[mon], batch_size=128, verbose=1)
y_pred = mdn.predict(s)
alpha_pred, mu_pred, sigma_pred = split_prediction(y_pred)
```

```python
from matplotlib import lines

plt.figure(figsize=(20, 10))

plt.title('MDN', fontsize=48, fontname='Times New Roman')

plt.plot(x_test,y_test,'x',alpha=.5, color="black", markersize=2)

for mx in range(mixtures):

    for i in range(len(mu_pred)):
        plt.errorbar(s[i],mu_pred[i,mx],
                        yerr=sigma_pred[i,mx],
                        alpha=alpha_pred[i,mx], 
                        color=tableau20[mx])
        
        plt.plot(s[i],mu_pred[i,mx], color=tableau20[mx],linewidth=1, marker='.', 
                linestyle='',alpha=alpha_pred[i,mx], markersize=10)# label='gaus_'+str(mx)+" prediction", )
        
ax = plt.gca()

ax.set_ylim(-140, 140)
ax.set_xlim(-10, 10)

#plt.axvline(x=5, color="black", alpha=0.7, linestyle="--")
#plt.axvline(x=9, color="black", alpha=0.7, linestyle="--")

#plt.axis('off')
plt.tight_layout()

plt.title(r'$y = \pm x^2 + \epsilon$', fontsize=32)

#plt.savefig("Multimod_Dist.png", format='png', dpi=300, bbox_inches='tight')

plt.show()
```
